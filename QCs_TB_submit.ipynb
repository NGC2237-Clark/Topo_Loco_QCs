```python
import matplotlib.patches as patches
def plot_shapes(shapes, title, save_path=None):
    """
    Plots a list of shapes using matplotlib.
    
    Args:
    - shapes (list): A list of shapes, where each shape is a list of (x, y) coordinates.
    - title (str): The title of the plot.
    """
    fig, ax = plt.subplots(figsize=(5,5))
    tick_label_fontsize = 12 
    label_fontsize = 20
    ax.set_title(title, fontsize=label_fontsize)


    for shape in shapes:
        polygon = patches.Polygon(shape, closed=True, fill=None, edgecolor='r')
        ax.add_patch(polygon)
    
    ax.set_aspect('equal')
    ax.autoscale_view()
    
    # Set tick label font size
    plt.xticks(fontsize=tick_label_fontsize)
    plt.yticks(fontsize=tick_label_fontsize)
    
    # Set axis label font size
    ax.set_xlabel('X', fontsize=label_fontsize)
    ax.set_ylabel('Y', fontsize=label_fontsize)
    plt.tight_layout()
    if save_path:
        fig.savefig(save_path, format='pdf', bbox_inches='tight')
    plt.show()
```

```python
# Example usage:
all_tiles = unique_coords_3
hopping_parameter = 0.3
hopping_d_max = 1/(1+np.sqrt(2)) +0.1

# connectivity minimum cutoff calculated by the consine rule in the equilateral triangle with 45 degree, 67.5 degree and 67.5 degree.
hopping_d_min = 0.3171 
#hopping_d_min = 0
E_onsite = 0
```

```python
h_matrix = qc.hamiltonian(hopping_parameter, all_tiles, hopping_d_max, hopping_d_min, E_onsite)
print(h_matrix)

print("The connectivity (coupling) diagram is shown below (the edges between vertices have non-zero hopping parameters):")
save_path = "C:\\Users\\11952\\OneDrive\\桌面\\Project_MPhi\\Preliminary_coding\\Plots_thesis\\connectivity.pdf"
qc.plot_connectivity(all_tiles, hopping_parameter, hopping_d_min, hopping_d_max, E_onsite, save_path)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(h_matrix)
# Get the indices that would sort the eigenvalues in ascending order
sorted_indices = np.argsort(eigenvalues)

# Sort the eigenvalues using the indices
sorted_eigenvalues = eigenvalues[sorted_indices]

# Sort the eigenvectors using the indices
sorted_eigenvectors = eigenvectors[:, sorted_indices]

sorted_eigenvalues = sorted_eigenvalues.real
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
# Plot the sorted eigenvalues
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(sorted_eigenvalues, 'o', markersize=3)
ax.set_xlabel('State Index', fontsize=24)
ax.set_ylabel('Energy', fontsize=24)
# ax.set_xticks(np.arange(len(sorted_eigenvalues)))
# ax.set_xticklabels(np.arange(len(sorted_eigenvalues)), fontsize=18)
# ax.set_yticks(np.linspace(min(sorted_eigenvalues), max(sorted_eigenvalues), num=5))
# ax.set_yticklabels(np.round(np.linspace(min(sorted_eigenvalues), max(sorted_eigenvalues), num=5), 2), fontsize=18)
# ax.grid(True)

# Set fewer x-ticks to avoid overcrowding
num_xticks = 5
xticks = np.linspace(0, len(sorted_eigenvalues) - 1, num_xticks, dtype=int)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks, fontsize=18)

# Set y-ticks
ax.set_yticks(np.linspace(min(sorted_eigenvalues), max(sorted_eigenvalues), num=5))
ax.set_yticklabels(np.round(np.linspace(min(sorted_eigenvalues), max(sorted_eigenvalues), num=5), 2), fontsize=18)
ax.grid(True)

# Inset parameters
inset_x, inset_y, inset_width, inset_height = 0.3, 0.5, 0.4, 0.4  # Relative to the main plot
axins = inset_axes(ax, width="40%", height="40%", loc='upper left', bbox_to_anchor=(0.15, 0.55, inset_width, inset_height), bbox_transform=ax.transAxes)

# Plot the same data in the inset
axins.plot(sorted_eigenvalues, 'o', markersize=3)

# Set limits for the inset axes to zoom in on a specific part
xmin, xmax = 125, 140  
ymin, ymax = -0.1, 0.1

axins.set_xlim(xmin, xmax)
axins.set_ylim(ymin, ymax)
# Set tick parameters for the inset
axins.tick_params(axis='both', labelsize=14)
# Mark the inset with lower left and upper right corners
mark_inset(ax, axins, loc1=3, loc2=1, fc="none", ec="0.5")
plt.tight_layout()
plt.savefig("C:\\Users\\11952\\OneDrive\\桌面\\Project_MPhi\\Preliminary_coding\\Plots_thesis\\qc_spectrum.pdf")
plt.show()
```
